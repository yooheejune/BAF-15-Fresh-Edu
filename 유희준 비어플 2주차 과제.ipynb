```python
#00시~06시를 새벽/06시~12시를 아침/12시~18시를 낮/18시~00시를 저녁으로 분류해보자
# 시간대 분류
def classify_hour(hour):
    if 0 <= hour < 6:
        return 'Dawn'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

df['Daytime'] = df['Hour'].apply(classify_hour)
df
```

```python
sns.countplot(x='Daytime',data=df)
```
